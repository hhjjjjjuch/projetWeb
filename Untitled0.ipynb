```python

IMAGE_SIZE = [224, 224]

train_path = '/content/drive/My Drive/archive/data/data1/train'
valid_path = '/content/drive/My Drive/archive/data/data1/valid'
```

```python
from tensorflow.keras.applications import InceptionV3

# Charger InceptionV3 avec les poids pré-entraînés sur ImageNet et sans la couche top
model = InceptionV3(input_shape=IMAGE_SIZE + [3], weights='imagenet', include_top=False)
```

```python
for layer in model.layers:
    layer.trainable = False
```

```python
from glob import glob  # Assure-toi d'importer glob

folders = glob('/content/drive/My Drive/archive/data/data1/train/*')
```

```python
folders
```

```python
from tensorflow.keras.layers import Flatten  # Assure-toi d'importer Flatten

x = Flatten()(model.output)
```

```python
from tensorflow.keras.layers import Dense   # Assure-toi d'importer Dense

prediction = Dense(len(folders), activation='softmax')(x)

# Créer l'objet modèle
from tensorflow.keras.models import Model  # Assure-toi d'importer Model

modelf = Model(inputs=model.input, outputs=prediction)
```

```python
modelf.summary()
```

```python
modelf.compile(
  loss='categorical_crossentropy',
  optimizer='adam',
  metrics=['accuracy']
)
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator

train_datagen = ImageDataGenerator(rescale = 1./255,
                                   shear_range = 0.2,
                                   zoom_range = 0.2,
                                   horizontal_flip = True)

test_datagen = ImageDataGenerator(rescale = 1./255)
```

```python
training_set = train_datagen.flow_from_directory('/content/drive/My Drive/archive/data/data1/train',
                                                 target_size = (224, 224),
                                                 batch_size = 32,
                                                 class_mode = 'categorical')
```

```python
test_set = test_datagen.flow_from_directory('/content/drive/My Drive/archive/data/data1/valid',
                                            target_size = (224, 224),
                                            batch_size = 32,
                                            class_mode = 'categorical')
```

```python
r = modelf.fit(
    training_set,
    validation_data=test_set,
    epochs=50,
    steps_per_epoch=len(training_set),
    validation_steps=len(test_set)
)
```
